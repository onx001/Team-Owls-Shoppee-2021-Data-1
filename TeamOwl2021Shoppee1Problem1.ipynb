```python
import numpy as np
import pandas as pd
import json
```

```python
file = open('contacts.json','r')
data = json.load(file)
```

```python
df = pd.DataFrame(data)
orderIds = []
emails = []
phones = []
df.tail()
```

```python
df = df.sort_values(by=['OrderId'])

back = 0
temp = []

for row in df.itertuples():
    
    if back == 0:
        back = row.OrderId
        
        if row.OrderId:
            temp.append(row.Id)
        
    

    elif row.OrderId == back:
        temp.append(row.Id)
        
    else:
        if len(temp) > 1:
            orderIds.append(temp)
        temp = []
        back = 0
    
orderIds = orderIds[1:]

dfOrderIds = pd.DataFrame(orderIds)

orderIds = []

dfOrderIds.tail()
```

```python
df = df.sort_values(by=['Email'])

back = ''
temp = []

for row in df.itertuples():
    
    if back == '':
        back = row.Email
        
        if len(back)>0:
            
            temp.append(int(row.Id))
        

    elif row.Email == back:
        temp.append(int(row.Id))
        
    else:
        if len(temp) > 1:
            emails.append(temp)
        temp = []
        back = ''
    
    
emails = emails[1:]

dfEmails = pd.DataFrame(emails)

emails = []

dfEmails.tail()
```

```python
df = df.sort_values(by=['Phone'])


back = ''
temp = []

for row in df.itertuples():
    
    if back == '':
        back = row.Phone
        
        if len(back)>0:
            temp.append(row.Id)

    elif row.Phone == back:
        temp.append(row.Id)
        
    else:
        if len(temp) > 1:
            phones.append(temp)
        temp = []
        back = ''
        
        
        

phones = phones[1:]


dfPhones = pd.DataFrame(phones)


phones = []

dfPhones.tail()
    
```

```python
master = np.empty(shape = (500000,), dtype = list)

```

```python
NpOrderIds = dfOrderIds.to_numpy().astype(int)


for indices in NpOrderIds:
    
    #print(indices)
    
    for index in indices:
        
        if not (0 < index < 500000):
            
            index = 0
            
        else:
            
            indices = indices[indices > 0]
            
            master[index] = indices.tolist()
            master[index].sort()
            
            
print(master[477150])
```

```python
NpEmails = dfEmails.to_numpy().astype(int)


for indices in NpEmails:
    
    #print(indices)
    
    for index in indices:
        
        if not (0 < index < 500000):
            
            index = 0
            
        else:
            
            indices = indices[indices > 0]
            
            if not master[index]:
                master[index] = []
                
            
            
            master[index] = list(dict.fromkeys(master[index] + indices.tolist())) #remove duplicates
            master[index].sort()
            
print(master[477150])
            
```
