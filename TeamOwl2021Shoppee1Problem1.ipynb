```python
import numpy as np
import pandas as pd
import json
```

```python
file = open('contacts.json','r')
data = json.load(file)
```

```python
df = pd.DataFrame(data)
orderIds = []
emails = []
phones = []
df.tail()
```

```python
df = df.sort_values(by=['OrderId'])

back = 0
temp = []

for row in df.itertuples():
    
    if not back:
        
        back = row.OrderId
        temp = [row.Id]
        
    

    elif row.OrderId == back:
        temp.append(row.Id)
        
    else:
        if len(temp) > 1:
            orderIds.append(temp)
        temp = []
        back = 0
    
#orderIds = orderIds[1:]

dfOrderIds = pd.DataFrame(orderIds)

orderIds = []

dfOrderIds.tail()
```

```python
df = df.sort_values(by=['Email'])

back = ''
temp = []

for row in df.itertuples():
    
    if not back:
        
        back = row.Email
        temp = [row.Id]
        

    elif row.Email == back:
        temp.append(int(row.Id))
        
    else:
        if len(temp) > 1:
            emails.append(temp)
        temp = []
        back = ''
    
    

dfEmails = pd.DataFrame(emails)

emails = []

dfEmails.tail()
```

```python
df = df.sort_values(by=['Phone'])


back = ''
temp = []

for row in df.itertuples():
    
    if not back:
        back = row.Phone
        temp = [row.Id]

        
    elif row.Phone == back:
        temp.append(row.Id)
        
    else:
        if len(temp) > 1:
            phones.append(temp)
        temp = []
        back = ''
        
        
        



dfPhones = pd.DataFrame(phones)


phones = []

dfPhones.tail()
    
```
